```python
import turicreate as tc
```

```python
housing_data = tc.SFrame('home_data.sframe')
```

```python
housing_data
```

```python
tc.visualization.scatter(housing_data['sqft_living'], housing_data['price'])
```

```python
tc.config.set_num_gpus(1)
```

# Dividing data into training and testing data

```python
train_data,test_data = housing_data.random_split(0.8)
```

```python
train_data.show()
```

```python
test_data.show()
```

```python
tc.visualization.scatter(test_data['sqft_living'], test_data['price'])
```

# Build a model using regression
- Using square foot prediction model through linear regression

```python
model_sqft_v_price = tc.linear_regression.create(train_data, target = 'price', features=['sqft_living'])
```

# Using the model and evaluate it

```python
# using the test data
print(test_data['price'].mean())
```

```python
model_sqft_v_price.evaluate(test_data)
```

# Predictions

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(test_data['sqft_living'], test_data['price'], '.', 
         test_data['sqft_living'], model_sqft_v_price.predict(test_data), '-')
```

```python
model_sqft_v_price.coefficients
```

# Using other features of the data

```python
list_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode']
```

```python
model_complex = tc.linear_regression.create(train_data, target='price', features = list_features)
```

```python
print(model_complex.evaluate(test_data))
print(model_sqft_v_price.evaluate(test_data))
```
